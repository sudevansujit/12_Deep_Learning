```
#RGB Color formats
import cv2
print([x for x in dir(cv2) if x.startswith('COLOR_')])

import cv2
img = cv2.imread('logo.png')
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY )
cv2_imshow(gray_img)
cv2.waitKey(0)

yuv_img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
y, u, v = cv2.split(yuv_img)
cv2_imshow(y)
cv2_imshow( u)
cv2_imshow( v)
cv2.waitKey(0)
```
